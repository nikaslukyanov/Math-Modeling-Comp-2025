```python
import pandas as pd
pd.set_option('display.max_columns', 500)
def findWaterLvl(file_path):
    #returns a dictionary with key = year and value = len(12) array w/ index corresponding to month
    data = pd.read_csv(file_path)
    monthlyLvl = dict()
    for index, row in data.iterrows():
        currArr = []
        for i in range(1,13):
            currArr.append(row[i])
        year = row[0]
        monthlyLvl[year] = currArr
    newData = data.drop(columns='Year')
    totalAvg = newData.values.mean()
    totalMax = newData.values.max()
    totalMin = newData.values.min()
    return monthlyLvl, totalAvg, totalMax, totalMin

def monthlyVals(file_path):
    #returns a dictionary with key = year and value = len(12) array w/ index corresponding to month
    data = pd.read_csv(file_path)
    yearlyLvl = dict()
    for index, row in data.iterrows():
        currArr = []
        for i in range(1,13):
            currArr.append(row[i])
        year = row[0]
        yearlyLvl[year] = currArr
    return yearlyLvl


#ontario beginning of month
file_path_waterLvl = "./Great Lakes Data/LakeOntario_MonthlyMeanWaterLevels_1918to2024.csv"
file_path_niagraFlow = "./Great Lakes Data/NiagaraRiver_MonthlyFlow_1900to2020.csv"
Ontario_Level = pd.read_csv(file_path_waterLvl)
Niagra_Flow = pd.read_csv(file_path_niagraFlow)
df_merged = pd.merge(Ontario_Level, Niagra_Flow, on='Year')[[' Jan', 'Jan']]
display(pd.merge(Ontario_Level, Niagra_Flow, on='Year'))
```

```python
# LOOK AT THIS CELL (cell below is AI generated)
import numpy as np
import matplotlib.pyplot as plt
import math
import sys
sys.stdout = open('log.txt', 'w')
def objective_functions(H, Q):
   # f1 = H + 1
   # f2 = (H-2) ** 2
    #f3 = H + Q
    U_1 = 51388 * (0.53 * (H - 74.7) - 0.7 * Q ** 2) if H >= 74.7 else 0
    C_1 = 51388*(0.05*(H - 74.9) ** 2) if H >= 74.9 else 0
    f1 = -(U_1 - C_1)

    # f2: Yield function of people managing docks (near Montreal harbor)
    f2 = -np.exp(-10.26 * abs(H - 73.56)) + 3.45 * Q**2

    # f3: Yield function of environmentalists
    f3 = -np.exp(-0.76 * abs(H - 74.83))

    # f4: Yield function of property owners on the shores of Lake Ontario
    f4 = -np.exp(-14.79 * abs(H - 74.56) - 3.68 * abs(Q - 7256)**2)

    # f5: Yield function of recreational boaters and fishing boats
    f5 = -np.exp(-5.89 * abs(H - 74.56) - 2.74 * abs(Q - 7256)**2)

    # f6: Yield function of hydro-power generation companies
    f6 = -46.35 * H

    # Return all objectives as a numpy array
    return np.array([f1, f2, f3, f4, f5, f6])

# Initialize population
def create_population(pop_size, n_var):
    return np.random.rand(pop_size, n_var) * 10 - 5

# Non-dominated sorting function
def non_dominated_sorting(population):
    pop_size = len(population)
    fronts = [[]]
    domination_counts = np.zeros(pop_size, dtype=int)
    dominated_solutions = [[] for _ in range(pop_size)]
    ranks = np.zeros(pop_size, dtype=int)

    for i in range(pop_size):
        for j in range(pop_size):
            if i != j:  # Skip self-comparison
                obj_i = objective_functions(*population[i])
                obj_j = objective_functions(*population[j])
                if np.all(obj_i <= obj_j):  # i dominates j
                    dominated_solutions[i].append(j)
                elif np.all(obj_j <= obj_i):  # j dominates i
                    domination_counts[i] += 1
        # If no solutions dominate i, it's part of the first Pareto front
        if domination_counts[i] == 0:
            ranks[i] = 0
            fronts[0].append(i)
    i = 0
    while fronts[i]:
        next_front = []
        for p in fronts[i]:
            for q in dominated_solutions[p]:
                domination_counts[q] -= 1
                if domination_counts[q] == 0:
                    ranks[q] = i + 1
                    next_front.append(q)
        i += 1
        fronts.append(next_front)

    return ranks, fronts[:-1]

pop_size = 100
n_var = 2
population = create_population(pop_size, n_var)

# Parameters for Lake Ontario simulation
H_mean = 74  # Average water level in feet
H_std = 1 # Standard deviation in feet (adjust for variability)
Q_mean = 7000  # Average flow rate in cubic meters per second
Q_std = 500  # Standard deviation for flow rate (adjust for variability)

# Number of samples to simulate
pop_size = 100
np.random.seed(42)
# Simulate the population of H and Q
Lake_Ontario = pd.read_excel(r'./Problem_D_Great_Lakes.xlsx', header=6, index_col='Year', sheet_name = 'Lake Ontario')
Niagara_River = pd.read_excel(r'./Problem_D_Great_Lakes.xlsx', header=6, index_col='Year', sheet_name = 'Niagara River', na_values="---")
combined_values = [[a,b] for a,b in zip(Lake_Ontario.values.flatten(), Niagara_River.values.flatten())]
combined_values_clean = [pair for pair in combined_values if not ((np.isnan(pair[0])) or (np.isnan(pair[1])))]
display(combined_values_clean)

# Perform non-dominated sorting
ranks, fronts = non_dominated_sorting(df_merged.values)
display(ranks, fronts)
colors = plt.cm.viridis(np.linspace(0, 1, len(fronts)))
plt.figure(figsize=(10, 6))
for i, front in enumerate(fronts):
    #solutions = combined_values_clean[front]
    solutions = [combined_values_clean[idx] for idx in front]  # Use the indices from 'front' to access 'combined_values_clean'
    solutions = np.array(solutions)
    plt.scatter(solutions[:, 0], solutions[:, 1], color=colors[i], label=f'Front {i + 1}', alpha=0.7)
display(df_merged.iloc[fronts[0], :].describe())
plt.title('Pareto Fronts Visualization For All January Data from 1918 to 2020')
plt.xlabel('H (Water Level)')
plt.ylabel('Q (Water Flow)')
plt.legend()
plt.grid(True)
plt.show()
```
